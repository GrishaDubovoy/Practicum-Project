### Технологический процесс
Как золото получают из руды? Изучите внимательно стадии процесса.
Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

Опишем каждую стадию:
1. Флотация

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

2. Очистка

Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### Расчет эффективности обогащения

Для расчета эффективности обогащения `rougher.output.recovery`  напишем функцию, но сначала напомним формулу 
 
где:
*	C — доля золота в концентрате после флотации/очистки - `rougher.output.concentrate_au`;
*	F — доля золота в сырье/концентрате до флотации/очистки - `rougher.input.feed_au`;
*	T — доля золота в отвальных хвостах после флотации/очистки - `rougher.output.tail_au`.

```python
def recovery_calc(row):
    try:
        c = row['rougher.output.concentrate_au']
        f = row['rougher.input.feed_au']
        t = row['rougher.output.tail_au']
        row['rougher.output.recovery.calc'] = ((c * (f - t)) / (f * (c - t))) * 100
    except ZeroDivisionError:
        return row
    return row
```

```python
df = df.apply(recovery_calc, axis=1)
```

```python
df['rougher.output.recovery'].hist(bins = 'auto')
```

Изначальные значения эффективности флотации выглядят верными, нет значений выше 100, но есть некоторое количество нулевых значений

```python
df[df['rougher.output.recovery'] == 0]['rougher.output.recovery'].count()
```

Теперь сравним с расчетами

```python
plt.hist(df['rougher.output.recovery.calc'], bins='auto', alpha = 0.5, label='Посчитанные по формуле')
plt.hist(df['rougher.output.recovery'], bins='auto', alpha = 0.5, label='Исходные данные')
plt.title('Гистограмма эффективности обогащения')
plt.legend(loc='upper left')
plt.show()
```

По гистограмме кажется что в наборе данных значения эффективности посчитаны корректно, посмотрим на MAE между двумя величинами

```python
print(mean_absolute_error(df['rougher.output.recovery.calc'],df['rougher.output.recovery']))
```

Отклонение очень мало, близко к нулю. Можно с уверенностью говорить что данные эффективности флотации предоставленные в исходном наборе данных корректны, а значит рассчитаный нами столбец лишний. Удалим его.

### Изменение концентрации металлов (Au, Ag, Pb) на различных этапах очистки

Для отслеживания изменения концентрации металлов на различных этапах построим графики

```python
# значения концентрация на выходе процессов флотации, очистки и итоговые концентрации
in_process = ['rougher.input.feed',
           'rougher.output.concentrate',
           'primary_cleaner.output.concentrate',
           'final.output.concentrate']
# значения концентрация в хвостах после этапоав флотации, очистки и итоговые хвосты
tail_process = ['rougher.output.tail',
                'primary_cleaner.output.tail',
                 'secondary_cleaner.output.tail',    
                'final.output.tail']
# металлы отслеживаемые
metals = ['au', 'ag', 'pb']
colors = ['gold', 'silver', 'plum']

fig, axes = plt.subplots(1, len(in_process),figsize=(20,5))
fig.suptitle('Концентрации на выходе процессов', fontsize=20)
for process,ax in zip(in_process,axes):
    ax.set_title(process)
    plt.title(process)
    for metal,color in zip(metals,colors):
        ax.hist(df[process+'_'+metal], alpha = 0.5, label=metal, bins=150, color=color, density=True)
        ax.legend(loc='upper left')
plt.show()

fig, axes = plt.subplots(1, len(tail_process),figsize=(20,5))
fig.suptitle('Концентрации в хвостах процессов', fontsize=20)
for process,ax in zip(tail_process,axes):  
    ax.set_title(process)
    plt.title(process)
    for metal,color in zip(metals,colors):
        ax.hist(df[process+'_'+metal], alpha = 0.5, label=metal, bins=150, color=color, density=True)        
        ax.legend(loc='upper left')
plt.show()
```

У свинца в итоговои концентрате также постепенно растет концентрация.

В хвостах концентрация золота примерно на одном уровне не зависимо от стадии
Серебра больше всего в хвостах после флотации, в последующих отвалах концентрация падает.
Свинца становиться сильно больше в хвостах первой очистки и совсем мало в хвостах от флотации.

## Модель

Для оценки результата тех процесса используется метрика описанная в разделе 1.3. Напишем функцию для расчета метрики

### Функции для метрики качества

```python
# функция для расчета итогового sMAPE
def total_smape(rougher_smape, final_smape):
    return ((0.25 * rougher_smape) + (0.75 * final_smape))
# функция для расчета sMAPE после флотации или в финальном концентрате
def smape(target, predict):
    result = 0
    for i in range(len(target)):
        result += np.abs(target[i]-predict[i]) / ((np.abs(target[i]) + np.abs(predict[i]))/2)
    result = result * 100 * (1 / len(target))
    return result
def smape_score(target, predict):
    result = (1/len(target)) * np.sum(np.abs(predict-target) / ((np.abs(target) + np.abs(predict))/2)*100)
    return result
scorer = make_scorer(smape_score, greater_is_better=False)
```
